# **Logistic Regression**

# **ROC Curve of Logistic Regression**

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

y__lr_proba = logreg.predict_proba(xtest)[:, 1]

fpr, tpr, thresholds = roc_curve(ytest, y__lr_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')
plt.title('AUC-ROC Curve for LR')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid(alpha=0.5)
plt.show()
```

# **K-Fold Cross Validation For Different ML Models**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
import xgboost as xgb
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import KFold, cross_val_score
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification


X, y = make_classification(n_samples=100, n_features=20, random_state=42)


models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "Naive Bayes": GaussianNB(),
    "XGBoost": xgb.XGBClassifier(),
    "Gradient Boosting": GradientBoostingClassifier()
}


kf = KFold(n_splits=5, shuffle=True, random_state=42)

results = {name: cross_val_score(model, X, y, cv=kf, scoring="accuracy") for name, model in models.items()}

for name, scores in results.items():
    print(f"{name} Mean Accuracy: {scores.mean():.4f}")


plt.figure(figsize=(10, 6))
for name, scores in results.items():
    plt.plot(range(1, kf.get_n_splits() + 1), scores, marker="o", label=name)

plt.title("K-Fold Cross-Validation Scores for Different Models")
plt.xlabel("Fold")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.show()
```

# **Comparison of False Negative Rate Before & After SMOTE**





```python
import matplotlib.pyplot as plt
import numpy as np

models = ['LR', 'RF', 'DT', 'KNN', 'NB', 'XGB', 'GB']
TP_before = [214, 1017, 1039, 0, 0, 1041, 1040]
FN_before = [827, 24, 2, 1041, 1041, 0, 1]

TP_after = [17693, 18427, 18384, 18070, 17608, 18393, 18377]
FN_after = [701, 30, 10, 324, 786, 1, 17]


FNR_before = [fn / (tp + fn) for tp, fn in zip(TP_before, FN_before)]
FNR_after = [fn / (tp + fn) for tp, fn in zip(TP_after, FN_after)]

FNR_before_percent = [fnr * 100 for fnr in FNR_before]
FNR_after_percent = [fnr * 100 for fnr in FNR_after]

x = np.arange(len(models))
width = 0.35

plt.figure(figsize=(12, 7))
plt.bar(x - width / 2, FNR_before_percent, width, label='Before SMOTE', color='skyblue', edgecolor='black')
plt.bar(x + width / 2, FNR_after_percent, width, label='After SMOTE', color='lightgreen', edgecolor='black')

plt.xlabel('Models', fontsize=12)
plt.ylabel('False Negative Rate (FNR) (%)', fontsize=12)
plt.title('Comparison of False Negative Rate (FNR) Before and After SMOTE', fontsize=14)
plt.xticks(x, models)
plt.ylim(0, max(max(FNR_before_percent), max(FNR_after_percent)) + 5)


for i, (before, after) in enumerate(zip(FNR_before_percent, FNR_after_percent)):
    plt.text(i - width / 2, before + 0.5, f'{before:.2f}%', ha='center', fontsize=10)
    plt.text(i + width / 2, after + 0.5, f'{after:.2f}%', ha='center', fontsize=10)

plt.legend(fontsize=12)


plt.tight_layout()
plt.show()
```
